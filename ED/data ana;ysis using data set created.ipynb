```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name": ["Adithya","Anil","Rohan","Anju","Arjun","Sneha","Vikram","Ananya","Kiran","Parvathy",
             "Suman","Neha","Nidhi","Divya","Manoj","Gokul","Raj","Priya","Nikhil","Ishan"],
    
    "Age": [25, 30, 28, 35, 40, 27, 45, 32, 29, 31,
            38, 26, 42, 33, 37, 24, 41, 34, 36, 28],
    
    "Department": ["IT","IT","Finance","IT","Management","HR","Finance","IT","HR","Finance",
                   "HR","IT","Finance","HR","Management","IT","Finance","HR","Management","HR"],
    
    "Salary": [80000, 50000, 45600, 60000, 82000, 32000, 70000, 55000, 35000, 48000,
               75000, 52000, 60000, 36000, 77000, 51000, 69000, 37000, 79000, 53000],
    
    "Experience": [6,5,4,7,10,3,9,6,3,4,
                   8,5,9,3,8,4,9,3,10,5],
    
    "PerformanceScore": [85,90,88,92,95,80,89,91,84,87,
                         93,86,90,82,94,85,91,83,96,88]
}

df = pd.DataFrame(data)
print(df)


```

```python
salary_array = np.array(df["Salary"])
exp_array = np.array(df["Experience"])

```

```python
print("Mean Salary:", np.mean(salary_array))
print("Median Salary:", np.median(salary_array))
print("Standard Deviation:", np.std(salary_array))
```

```python
print("Max Experience:", np.max(exp_array))
print("Min Experience:", np.min(exp_array))

# Normalize Salary
normalized_salary = (salary_array - np.min(salary_array)) / (np.max(salary_array) -np.min(salary_array))
print("Normalized Salary:", normalized_salary)
```

```python



print("\nFirst 5 rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())


avg_salary = df.groupby("Department")["Salary"].mean()
print("\nAverage Salary by Department:\n", avg_salary)


top_employee = df.loc[df["PerformanceScore"].idxmax()]
print("\nTop Performer:\n", top_employee)


df["Salary_After_Bonus"] = df["Salary"] * 1.10
print("\nUpdated DataFrame:\n", df.head())

```

```python

plt.figure()
avg_salary.plot(kind='bar')
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
plt.subplot(2, 2, 2)
df_sorted = df.sort_values('Age')
plt.plot(df_sorted['Age'], df_sorted['Salary'], marker='o', linestyle='-', color='green')
plt.title('Age vs. Salary Trend')
plt.xlabel('Age')
plt.ylabel('Salary ($)')
plt.grid(True, alpha=0.3)
```

```python
plt.figure()
df["Department"].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Department Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.subplot(2, 2, 4)
plt.hist(df['Salary'], bins=6, color='orange', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary Range')
plt.ylabel('Number of Employees')

plt.tight_layout()
plt.show()
```
